```python
#Plot a graph by multiplotting on the same canvas (Take any set of x & y values)

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Create some data
Yearly_Outcomes =  pd.read_excel("OrganDonor_Consolidated.xlsx", sheet_name='Yearly_Outcomes') 
x=Yearly_Outcomes['OPO']
y=Yearly_Outcomes['Year']
z=Yearly_Outcomes['mean calc deaths']

# Create a figure and a set of subplots
fig, axs = plt.subplots(2, 1) 

# Plot the first subplot
axs[0].plot(x, y)
axs[0].set_title('Age and cuaseofdeath')

# Plot the second subplot
axs[1].plot(x, z)
axs[1].set_title('age and mechanism_of_death')

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Example dataset
data = {
    'Age': [22, 30, 37, 45, 52, 65, 73, 25, 39, 60],
    'Cause of Death': ['Accident', 'Disease', 'Natural Causes', 'Disease', 'Accident',
                       'Disease', 'Natural Causes', 'Accident', 'Other', 'Other']
}
df = pd.DataFrame(data)

# Step 1: Create age groups
bins = [0, 25, 35, 45, 55, 65, 75]
labels = ['18-25', '26-35', '36-45', '46-55', '56-65', '66-75']
df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

# Step 2: Create a pivot table to count donors by age group and cause of death
pivot_table = df.pivot_table(index='Age Group', columns='Cause of Death', aggfunc='size', fill_value=0)

# Step 3: Calculate the correlation matrix
correlation_matrix = pivot_table.corr()
correlation_matrix 
```
